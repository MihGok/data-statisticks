```python
data  = {'height':[1.735,1.795,1.832,1.86,1.92,1.772,1.97,1.87,1.59,1.79,2.03],'month':[5,10,9,1,11,9,5,11,5,10,10],'number':[8,1,6,0,7,8,1,3,7,2,1]}
```

```python
def stats(ds):
  ds = sorted(ds)
  avg = sum(ds)/len(ds)
  print('average = ',avg)
  arr1 = [(elem - avg)**2 for elem in ds]
  disp = sum(arr1) / len(arr1)
  print('dispersion = ',disp)
  standart = disp**0.5
  print('standart = ', standart)
  if len(ds) %2 == 1:
    median = ds[len(ds)//2]
  else:
    median = 0.5*(ds[len(ds)//2] + ds[len(ds)//2 - 1])
  print('median = ',median,ds,len(ds))
  variation = standart / avg
  print('variation coef = ',variation)
```

```python
for i in range(len(data.keys())):
  print('for ',list(data.keys())[i],' statistics is')
  stats(data[list(data.keys())[i]])
  print('\n\n')
```

```python
month_unique,month_count = np.unique(data['month'], return_counts = True)
print(f'Unique values for month are {month_unique}')
print(f'Their absolute frequency are {month_count}')
print(f'Their relative frequency are {month_count/sum(month_count)}')
```

```python
month_unique, month_count = np.unique(data['month'], return_counts=True)
df = pd.DataFrame({
    'Unique Values': month_unique,
    'Absolute Count': month_count,
    'Relative Count': month_count / sum(month_count)
})

print(df.T)
plt.plot(df['Unique Values'], df['Relative Count'],color = 'red',marker = 'o', linestyle = ':')
plt.title('Полигон относительных частот для месяца')
plt.xlabel('Значения')
plt.ylabel('Относительные частоты')
```

```python
df['Cumulative Relative Count'] = df['Relative Count'].cumsum()
print(df.T)


plt.figure(figsize=(10, 6))
plt.step(df['Unique Values'], df['Cumulative Relative Count'], color = 'blue',marker = 'o', linestyle = '-')
plt.title('Function for month')
plt.xlabel('Month')
plt.ylabel('F(Month)')
plt.grid(True)
```

```python
def calculate_intervals(data):
    n = len(data)
    m = round(1 + 3.322 * math.log10(n))
    return m

def calculate_bounds(num, h, min):
  start = min - h/2
  end = min + h/2
  inter = [[start,end]]
  for _ in range(num):
      start,end = end, end + h
      inter.append([start,end])
  return np.array(inter)

m = calculate_intervals(data['height'])
h = (max(data['height'])-min(data['height']))/m
print(f"Число интервалов (m): {m}\n")
print(f"Длина интервалов  {h}")
bounds = calculate_bounds(m,h,min(data['height']))
print(f"Границы интервалов\n {calculate_bounds(m,h,min(data['height']))}\n")

bin_edges = np.unique(calculate_bounds(m,h,min(data['height'])))
bin_indices = np.digitize(data['height'], bin_edges)
print(data['height'])

interval_bins = {i: [] for i in range(1, len(bin_edges))}
for i, value in zip(bin_indices, data['height']):
    if i < len(bin_edges):
        interval_bins[i].append(value)

for i, values in interval_bins.items():
    print(f"Интервал {bin_edges[i-1]} - {bin_edges[i]}: {values}")
```

```python
data1 = pd.DataFrame(
    {
        "Number":data['height'],
        "Belongs to":bin_indices,
    }
)
```

```python
def get_string_bounds(bounds):
  res = []
  for elem in bounds:
    res.append(str(round(elem[0],3)) + '-' + str(round(elem[1],3)))
  return res

bounds_str = get_string_bounds(bounds)
print(bounds_str)
```

```python
print(f"Variational series {sorted(bin_indices)}")
```

```python
interval_counts = data1['Belongs to'].value_counts().sort_index()
relative_frequencies = interval_counts / len(data1)
data2 = pd.DataFrame(
    {
        'Interval':[i for i in range(1,6)],
        'Interval absolute frequency':interval_counts,
        'Interval relatuve frequency':relative_frequencies,
        'Bounds':bounds_str
    }

)
print(data2.T)
```

```python
plt.figure(figsize=(10, 6))
plt.title('Interval distribution')
plt.xlabel('Number')
plt.ylabel('F(number)')
plt.grid(True)
plt.bar(data2['Bounds'], data2['Interval absolute frequency'], color = 'grey', linestyle = '-')
```

```python
data2['cumulative_frequencies'] = data2['Interval relatuve frequency'].cumsum()
print(data2.T)


plt.figure(figsize=(10, 6))
plt.step(data2['Interval'], data2['cumulative_frequencies'], color = 'orange',marker = 'o', linestyle = '-')
plt.title('Function for number')
plt.xlabel('Number')
plt.ylabel('F(number)')
plt.grid(True)
```
